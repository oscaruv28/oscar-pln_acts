# **Script guia para la solución de la actividad 10**



1. **Carga y exploración de datasets:**
Cargar el dataset TASS de manera local para el análisis de sentimientos, mostrando la distribución de etiquetas en los tres conjuntos de datos (entrenamiento, validación y prueba). Además, descargar el dataset de Sarcasmo directamente desde Hugging Face.

2. **Ajuste fino (Fine-tuning) de modelos de clasificación:**
- Sobre el dataset TASS: realizar el fine-tuning de los modelos XLNet y XLM-RoBERTa-large para la clasificación de tres sentimientos (positivo, neutro y negativo).
- Sobre el dataset de Sarcasmo: afinar los modelos BETO y XLM-RoBERTa-large para la clasificación binaria (sarcasmo vs. no sarcasmo).

3. **Evaluación de métricas de desempeño (TASS y Sarcasmo):**
Calcular las métricas de Recall, Precisión y F1-score para los modelos afinados en ambos datasets. Las pruebas deben ejecutarse configurando batch sizes de 8, 16 y 32, con un rango de 6 a 10 épocas de entrenamiento.

4. **Script de inferencia con modelos de Hugging Face:**
Desarrollar un script para realizar pruebas de inferencia utilizando los mejores modelos publicados en Hugging Face, evaluando su respuesta ante varios ejemplos de prueba de ambos datasets.

5. **Análisis comparativo de resultados:**
Realizar un análisis crítico de los dos procesos de afinamiento, respondendo a las siguientes preguntas:
- ¿Por qué los rendimientos sobre el dataset TASS son menores que los obtenidos sobre el dataset de Sarcasmo?
- ¿Qué papel juega el número de clases en el desempeño de ambos datasets?
- Comparar el desempeño de los modelos XLNet y XLM-RoBERTa-large (en TASS) frente a BETO y XLM-RoBERTa-large (en Sarcasmo).


6. **Entrenamiento de modelos BiLSTM para NER:**
Entrenar dos modelos de BiLSTM para resolver la tarea de Reconocimiento de Entidades Nombradas (NER) sobre el dataset CoNLL-2002. El objetivo es superar el F1-score del modelo base BiLSTM, que actualmente es del 64.3%.

7. **Afinamiento de LLMs para el dominio médico:**
Realizar el fine-tuning de los modelos BETO y XLM-RoBERTa sobre el dataset de cáncer de próstata, respetando la parametrización definida para LLMs. Se deben probar batch sizes de 8, 16 y 32, y aplicar técnicas de optimización/cuantización (como LoRA, QLoRA o XLora, según la especificación del curso) si es necesario.


8. **Despliegue de modelos en Hugging Face:**
Subir los mejores modelos afinados a la plataforma Hugging Face y desarrollar un script para probar su inferencia con diversos textos relacionados con el cáncer de próstata.

9. **Evaluación de métricas para el dataset de próstata:**
Calcular las métricas de desempeño (Recall, Precisión y F1-score) específicamente para el fine-tuning de los modelos sobre el dataset de próstata, evaluando configuraciones con batch sizes de 8, 16 y 32, y un rango de 6 a 10 épocas.







10. **NER biomédico mediante Prompt Engineering:**
Desarrollar un script para resolver la tarea de NER biomédico utilizando Prompt Engineering con LLMs generativos de código abierto, tales como Mistral, Llama, Gemma y DeepSeek.
